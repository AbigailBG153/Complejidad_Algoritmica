#Leyndo imput as a file

```python
def solve(m,n,m0,n0,m1,n1):
  mc = m0 + m1
  nc = n0 + n1
  if mc == m and nc == n:
    return ""
  if m / n < mc/nc:
    return f"L{solve(m,n,m0,n0,mc,nc)}"
  else:
    return f"R{solve(m,n,mc,nc,m1,n1)}"     
```

```python
with open("1.in") as f:
  for line in f :
    m,n = [int(x) for x in line.split()]
    if m == 1 and n==1:
       break 

    resp = solve(m,n,0,1,1,0)
    print(resp)   
```

Reading input as stdin

```python
while True :
    m,n = [int(x) for x in input().split()]
    if m == 1 and n==1:
       break 

    resp = solve(m,n,0,1,1,0)
    print(resp)   
```
