```python
import numpy as np
import matplotlib.pyplot as plt
def drawBoard(board):
   n = len(board)
   b = np.zeros((n,n), dtype=int) 
   b[::2,::2] = 1
   b[1::2,1::2] = 1
   _, ax = plt.subplots()
   ax.imshow(b)
   plt.imshow(b)
   for row,col in enumerate(board):
    ax.text(col,row,"Q", fontsize = 150/n , va= "center" , ha = "center" )
   
```

```python
def isCorrect(board,row,col):
  for rowi in range(row):
    delta = row - rowi
    colright = board[rowi]+ delta
    colleft = board[rowi]- delta
    if board[row] == col or col == colright or col == colleft:
      return False
  return True
```

```python
def nqueen(board,row):
  n = len(board)
  if row == n:
    drawBoard(board)
  else:
    for col in range(n):
      if isCorrect(board,row,col):
         board[row] = col
         nqueen(board,row + 1)
         board[row] = -1  
```

```python
n=4
nqueen([-1]*n , 0, n)
```
