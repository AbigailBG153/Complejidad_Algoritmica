```python
def Swap(a,b):
  a,b = b,a
```

```python
def Combinacion( G,k,n):
  if k==n:
      print(G)
  else:
    for i in range(k,n):
        Swap(G[i],G[k])
        Combinacion(G,k+1,n)
        Swap(G[i],G[k])

```

```python
a = ['A','B','C']
j = len(a)
for i in range(j+1):
 print("Con",i,"Combinaciones : ")
 print(Combinacion(a,0,2))
```
