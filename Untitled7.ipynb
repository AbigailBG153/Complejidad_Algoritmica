##DFS

```python
def dfs(G):
  n = len(G)
  visited = [False]*n
  lst = []

  def d(u):
    visited[u] = True
    for v in G[u]:
      if not visited[v]:
        d(v)

    lst.append(u)
  
  d(0)
  return len(lst)      

```

```python
with open("1.in") as f:
  t=0
  while True:
    t += 1
    n,m = map(int,f.readline().split())
    if n == 0 and  m ==0 :
      break
    G = [[] for _ in range(n)]
    for i in range(m):
      nums = [int(n) for n in f.readline().split()]
      k = nums[0]
      nums = nums[1:]
      for i in range(k-1):
        u = nums[i]
        v= nums[i+1]
        G[u].append(v)
        G[v].append(u)

    print(f"# Sospechosos - Caso #{t}: {dfs(G)}")    
```
