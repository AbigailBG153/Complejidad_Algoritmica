```python
import heapq as hq
import math

def difkstra(G,s):
  n = len(G)
  visited = [False]*n
  path=[-1]*n
  g=[math.inf]*n
  g[s]=0

  pq=[]
  hq.heappush(pq,(0,s))
  while len(pq) > 0 : # mientras que haya algun elemneto en la cola de prioridad
    _, u= hq.heappush(pq)
    visited[u]=True
    for v , w in G[u]:
      f = g[u]+ w
      if not visited[v] and f < g[v]:
        g[v] = f
        path[v] = u
        hq.heappush(pq,(f,v))

  return path,g      
```

```python
def dls(G,s,L):
  n = len(G)
  visited = [False]* n
  path = [-1]*n

  def _f(u,l):
    if l>0:
      visited[u] = True
      for v in G[u]:
        if not visited[v]:
          path[v] = u
          _f(v,l-1)

  _f(s,L)
  return path        
```
