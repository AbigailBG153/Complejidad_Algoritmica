Partir numeros en dos 


```python
def dvmult(a, b, n):
  if n == 1:
    return a * b

  ai = a // 10**(n//2)
  ad = a % 10**(n//2)
  bi = b // 10**(n//2)
  bd = b % 10**(n//2)
  z1 = dvmult(ai, bi, n//2) * 10**n
  z2 = (dvmult(ai, bd, n//2) + dvmult(ad, bi, n//2)) * 10**(n//2)
  z3 = dvmult(ad, bd, n//2)
  
  return z1 + z2 + z3
```

```python
def multi(a,b):
  m = max(len(str(a)),len(str(b)))
  
  return dvmult(a,b,m)
```

```python
a = 548929495
b = 102043934
res = multi(a, b)
print(res)
"""assert res == a * b"""

```
